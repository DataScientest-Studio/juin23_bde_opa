```python
import json

from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

```python
stub_data = "/home/theenglishway/Documents/datascientest/JUIN23-BDE-OPA/data/examples/fmpcloud_AAPL_history.json"
with open(stub_data) as f:
    data = f.read()
    as_json = json.loads(data)

df = pd.DataFrame([d for d in as_json["historical"]])
df
```

```python
app = Dash(__name__)


def add_range_selectors(figure):
    return figure.update_xaxes(
        rangeslider_visible=True,
        rangeselector=dict(
            buttons=list(
                [
                    dict(count=1, label="1m", step="month", stepmode="backward"),
                    dict(count=6, label="6m", step="month", stepmode="backward"),
                    dict(count=1, label="YTD", step="year", stepmode="todate"),
                    dict(count=1, label="1y", step="year", stepmode="backward"),
                    dict(step="all"),
                ]
            )
        ),
    )


def into_candlestick(df):
    return go.Candlestick(
        x=df.date, open=df["open"], high=df["high"], low=df["low"], close=df["close"]
    )


app.layout = html.Div(
    [
        html.H1(children="Stock price (AAPL)", style={"textAlign": "center"}),
        dcc.Graph(
            figure=add_range_selectors(go.Figure(into_candlestick(df))),
            id="stock-value",
        )
    ]
)


app.run()
```

```python
app.layout = html.Div(
    [
        html.H1(children="Volume of stock exchange (AAPL)", style={"textAlign": "center"}),
        dcc.Graph(
            figure=add_range_selectors(px.bar(df, x="date", y="volume")), id="volume"
        ),
    ]
)


app.run()
```
